```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
data = pd.read_csv('data1.csv')
```

```python
def load_dataset(filename):
    data = pd.read_csv(filename)
    dataset = data.values
    #input
    X = dataset[:,:-1]
    #the target variable
    y = dataset[:,-1]
    print(X, "\n Y is ", y)
    return X,y
```

```python
def prepare_input(X):
    ohe = OneHotEncoder()
    X_trans_enc = ohe.fit_transform(X)
    return X_trans_enc
```

```python
def prepare_targets(y):
    le=LabelEncoder()
    #le.fit(y_train)
    y_trans_enc = le.fit_transform(y)
    #print(y_train_enc)
    #y_test_enc = le.fit_transform(y_test)
    return y_trans_enc
```

```python
X, y = load_dataset('data1.csv')
```

```python
#input data preparation
X_trans_enc = prepare_input(X)
print(X_trans_enc)
```

```python
#target data preparation
y_trans_enc = prepare_targets(y)
```

```python
# Split the dataset into training and testing
X_train, X_test, y_train, y_test = train_test_split(X_trans_enc, y_trans_enc, test_size=0.2, random_state = 1)
X_val, test_set, y_val, y_test_set = train_test_split(X_test, y_test, test_size=0.5)
```
